Tranformada Discreta de fourier sobre una imagen

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import fft2, fftshift, ifft2
```

Vamos a crear una imagen de escala de grises con valores periodicos

Creamos el espacio de los valos de x, y de acuerdo a las dimensiones dadas de filas y columnas

```python
width2 = 600
heigh2 = 300

x = np.linspace(0,2*width2/(2*width2+1),2*width2+1)
y = np.linspace(0,2*heigh2/(2*heigh2+1),2*heigh2+1)
```

Creamos la malla de la imagen

```python
[X,Y] = np.meshgrid(x,y)
```

Definimos la frecuencia de la funcion

```python
frequencyW = 10.1
frequencyH = 20.5
```

creamos nuestra funcion de x,y

```python
functionA = np.sin(2*np.pi*frequencyW*X)
```

Graficamos la Imagen

```python
axisX = np.linspace(0,1,1200)
plt.imshow(functionA, extent=[0,1,0,1], cmap='gray')
plt.show
```

Relizamos la tranformada de fourier

```python
heigh,width = np.shape(functionA)
funtionFA = fft2(functionA)/(heigh*width)
funtionFA = fftshift(funtionFA)
potFA = np.abs(funtionFA)
plt.imshow(potFA, extent=[-width2,width2,-heigh2,heigh2])
plt.show()
```

```python
plt.imshow(potFA[heigh2-25:heigh2+25,width2-25:width2+25], extent=[-25,25,-25,25])
plt.show()
```
