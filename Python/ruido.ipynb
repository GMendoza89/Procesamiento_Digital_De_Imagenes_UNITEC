Ejercicio vamos a crear una función que genere el ruido de sal y pimiento

primer paso:
Importar los modulos necesarios
 OpenCV
 Numpy
 ramdom - generar numeros aleatorios
 metplotlib.pyplot

```python
import cv2 as cv
import numpy as np
import random as rd
import matplotlib.pyplot as plt
```

definieremos la funcion que incorpore el ruido a la imagen

entrada: imagen
salida: imagen con ruido de sal y pimienta

```python
def addNoise(image):
    rows, cols = image.shape
    #factorNoise = rd.random()
    factorNoise = 0.15
    numberOfPixels = int(image.size*factorNoise)
    for pixel in range(0,numberOfPixels):
        rdRow = rd.randint(0,rows - 1)
        rdCol = rd.randint(0,cols - 1)
        if rd.random()>0.5:
            image[rdRow,rdCol] = 255
        else:
            image[rdRow,rdCol] = 0

```

haremos la lectura de la imagen

```python
img = cv.imread('lenna.jpeg',0)
```

Generar una copia de nuesta imagen

```python
imgNoise = img.copy()
```

mandamos a llamar nuestra función que nos genera el ruido de sal y pimienta

```python
addNoise(imgNoise)
```

graficamos las dos imagenes

```python
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('Imagen de entrada')
plt.xticks([])
plt.yticks([])
plt.subplot(122)
plt.imshow(imgNoise, cmap='gray')
plt.title('Imagen con ruido')
plt.xticks([])
plt.yticks([])
plt.show()
```
