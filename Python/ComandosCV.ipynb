Se presentaran calgunos comandos útiles en OpenCV y Python

iniciaremos con la craga de imagenes

```python
import cv2 as cv # Libreria para manejo de imagen
import numpy as np# Librería para manejo de matrices
import matplotlib.pyplot as plt

img = cv.imread('lenna.jpeg',1)
imgGray = cv.imread('lenna.jpeg',0)

```

adquirir informacion de dimensiones de la imgen tenemos shape que nos entrega una tupla de el largo y el ancho de la imagen, y la propiedad size nos entrega un valor con el total de pixeles de la imagen


```python
dimension = img[:,:,0].shape
pixelsSize = img[:,:,0].size

print(dimension)
print(pixelsSize)
```

Crear un diccioneario de colores
para crear colores en opencv
Azul -> 255, 0, 0
dark_gray -> 50,50,50
light_gray -> 220,220,220

```python
colors = {'blue':(255,0,0),'green':(0,255,0),'red':(0,0,255),'yellow':(0,255,255),'darkGray':(20,20,20),'lightGray':(220,220,220),'white':(255,255,255),'black':(0,0,0)}
```

Dibujo de formas en Python y OpenCV
primero se creara una imagen de 400 x 400 pixeles

```python


imgP = np.zeros((400,400,3),np.uint8)
imgP[:] = colors['lightGray']

imgPlot = imgP[:,:,::-1]
plt.imshow(imgPlot)
#plt.show()
```

Dibuajndo una linea

```python
cv.line(imgP,(0,50),(400,50),colors['blue'],10)
cv.line(imgP,(0,0),(400,400),colors['yellow'],5)
imgPlot = imgP[:,:,::-1]
plt.imshow(imgPlot)
#plt.show()
```

Para dibujar un rectangulo en una figura
la funcion que se encarga el la rectangle y recibe 5 cinco paramtros
 * La imagen a donde dibujar
 * punto incial de la imagen -> esquina superior izquierda
 * puntofinal ->esquina inferior derecha
 * color (bgr)
 * ancho del borde, si el es -1 genera una cuadro solido


```python
imgP = np.zeros((400,400,3),np.uint8)
imgP[:] = colors['lightGray']
cv.rectangle(imgP, (10,10),(100,100),colors['red'],1)
cv.rectangle(imgP, (50,500),(250,100),colors['blue'],-1)
imgPlot = imgP[:,:,::-1]
plt.imshow(imgPlot)
```

Para dibujar un circulo
se utiliza la función circle
* La imagen sobre la cual dibujar
* la coordenadas del centro del circulo
* radio del circulo
* el color (b,g,r)
* el grosor de la linea si es -1 en un circulo de color solido 

```python
imgP = np.zeros((400,400,3),np.uint8)
imgP[:] = colors['lightGray']
cv.circle(imgP,(199,199),100,colors['black'],1)
cv.circle(imgP,(199,199),50,colors['blue'],1)
cv.circle(imgP,(199,199),10,colors['red'],-1)
imgPlot = imgP[:,:,::-1]
plt.imshow(imgPlot)
```

Rotar una imagen


```python
rotationMatrix = cv.getRotationMatrix2D((dimension[0]/2,dimension[1]/2),45,0.70)
```

```python
imgRotate = cv.warpAffine(img,rotationMatrix,dimension)
```

```python
pltimgr = imgRotate[:,:,::-1]
plt.imshow(pltimgr)
```

Cambiar brillo y contraste en una imagen

new imagen = alpha*imagen original+beta

```python
contrastImage = cv.addWeighted(img,2.5,np.zeros(img.shape,img.dtype),0,0)

imgplot = img[:,:,::-1]
imgcontplot = contrastImage[:,:,::-1]
plt.subplot(121)
plt.imshow(imgplot)
plt.title("Imagen original")
plt.subplot(122)
plt.imshow(imgcontplot)
plt.title("Imagen Contrastada")
```

```python
kernelsize = 31
blurImage = cv.GaussianBlur(img,(kernelsize,kernelsize),0)

blurPlot = blurImage[:,:,::-1]
plt.subplot(121)
plt.imshow(imgplot)
plt.title("Imagen original")
plt.subplot(122)
plt.imshow(blurPlot)
plt.title("Blur image")
```

Filtro de media
Se utiliza principal mente para eliminar ruido de sal y pimienta

```python
kernelsize = 5
medianBlurImage = cv.medianBlur(img,kernelsize)

medianblurPlot = medianBlurImage[:,:,::-1]
plt.subplot(121)
plt.imshow(imgplot)
plt.title("Imagen original")
plt.subplot(122)
plt.imshow(medianblurPlot)
plt.title("Blur image")
```

```python
thres1 = 50
thres2 = 50

edgeimg = cv.Canny(img,thres1,thres1)

plt.subplot(121)
plt.imshow(imgplot)
plt.title("Imagen original")
plt.subplot(122)
plt.imshow(edgeimg,cmap='gray')
```
