Crear un diccioneario de colores
para crear colores en opencv
Azul -> 255, 0, 0
dark_gray -> 50,50,50
light_gray -> 220,220,220

```python
colors = {'blue':(255,0,0),'green':(0,255,0),'red':(0,0,255),'yellow':(0,255,255),'darkGray':(20,20,20),'lightGray':(220,220,220),'white':(255,255,255),'black':(0,0,0)}
```

Dibujo de formas en Python y OpenCV
primero se creara una imagen de 400 x 400 pixeles

```python


imgP = np.zeros((400,400,3),np.uint8)
imgP[:] = colors['lightGray']

imgPlot = imgP[:,:,::-1]
plt.imshow(imgPlot)
#plt.show()
```

Dibuajndo una linea

```python
cv.line(imgP,(0,50),(400,50),colors['blue'],10)
cv.line(imgP,(0,0),(400,400),colors['yellow'],5)
imgPlot = imgP[:,:,::-1]
plt.imshow(imgPlot)
#plt.show()
```

Para dibujar un rectangulo en una figura
la funcion que se encarga el la rectangle y recibe 5 cinco paramtros
 * La imagen a donde dibujar
 * punto incial de la imagen -> esquina superior izquierda
 * puntofinal ->esquina inferior derecha
 * color (bgr)
 * ancho del borde, si el es -1 genera una cuadro solido


```python
imgP = np.zeros((400,400,3),np.uint8)
imgP[:] = colors['lightGray']
cv.rectangle(imgP, (10,10),(100,100),colors['red'],1)
cv.rectangle(imgP, (50,500),(250,100),colors['blue'],-1)
imgPlot = imgP[:,:,::-1]
plt.imshow(imgPlot)
```

Para dibujar un circulo
se utiliza la función circle
* La imagen sobre la cual dibujar
* la coordenadas del centro del circulo
* radio del circulo
* el color (b,g,r)
* el grosor de la linea si es -1 en un circulo de color solido 

```python
imgP = np.zeros((400,400,3),np.uint8)
imgP[:] = colors['lightGray']
cv.circle(imgP,(199,199),100,colors['black'],1)
cv.circle(imgP,(199,199),50,colors['blue'],1)
cv.circle(imgP,(199,199),10,colors['red'],-1)
imgPlot = imgP[:,:,::-1]
plt.imshow(imgPlot)
```
